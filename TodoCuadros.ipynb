```python
lista_info_cuad=list()
```

```python
for i in pintores:
    #Creamos carpeta para el pintor y entramos en ella
    navegador.get("https://www.todocuadros.es/") 
#     os.mkdir(i)
#     os.chdir(i)
    
    #buscamos el pintor en el navegador
    sleep(randint(1,1))
    busq = navegador.find_element_by_id("search")
    busq.send_keys(i)
    sleep(randint(1,1))
    busq.submit()
    
    #soup de la página actual del navegador
    source = navegador.page_source
    soup = BeautifulSoup(source)

        
    #Recorre la lista de cuadros y guarda cada imagen
    for j in range(len(soup.find_all("span", class_="snize-overhidden"))):
        cuad=(soup.find_all("span", class_="snize-overhidden")[j].find("span").text)
        print(cuad)
        sleep(randint(1,1))
        navegador.find_element_by_partial_link_text(cuad).click()
        soup = BeautifulSoup(navegador.page_source)
        #imagen del cuadro
#         try:
#             img_cuadro = soup.find("div", class_="magic-slide mt-active").find("a")["href"]
#             resp = requests.get(img_cuadro, stream=True)
#             imagen = open(cuad+'.jpg', 'wb')
#             resp.raw.decode_content = True
#             shutil.copyfileobj(resp.raw, imagen)
#             imagen.close()
#         except:
#             print("no hay imagen")
            
        #buscar la info del cuadro
        try:
            info_cuad=soup.find("div", class_="relacionados")
            info_cuad=info_cuad.find_all("div", class_="item-relacionado")
            claves=list()
            valores=list()
            #guardar en un diccionario por claves y valores la info del cuadro
            for i in info_cuad:
                claves.append(i.text.split(":")[0])
                valores.append(i.text.split(":")[1])

            #creamos una lista de diccionarios con la info de cada cuadro
            dic_aux=dict()
            for key,value in zip(claves,valores):
                dic_aux.update({key:value})

            lista_info_cuad.append(dic_aux)
            print(dic_aux)
        except:
            print("nada")
        sleep(randint(1,1))
        navegador.execute_script("window.history.go(-1)")
        #Actualizamos la variable soup con la nueva página actual
        soup = BeautifulSoup(navegador.page_source)
        sleep(randint(1,1))
    
    #Create a DataFrame object
    df = pd.DataFrame()
    #Add new ROW
    for i in lista_info_cuad:
        df=df.append(i , ignore_index=True)
    

    #Volvemos al directorio "Imagenes"
#     os.chdir("..")
    break
```
